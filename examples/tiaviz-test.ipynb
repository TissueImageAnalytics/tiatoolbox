```python
from pathlib import Path
model = GeneralSegmentor(SAM())

glands = "slides/glands.png"
slides = "slides/sample_wsi.svs"
glands_prompts = [(370,270),(300,400)]
slides_prompts = [[5792,6018]]
slides_location = (5745,5972)
slides_size = (200,114)
```

```python
prompts = model.create_prompts(slides_prompts)
output = model.predict(slides, prompts, "cpu", "overlays", slides_location,slides_size, 0.5)

```
